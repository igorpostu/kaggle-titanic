# EDA

### Loading the packages

```python
import pandas as pd
import seaborn as sns
```

### Performing EDA

Loading the dataset. We need only the `train` data for the analysis.

```python
data = pd.read_csv("../data/train.csv")
```

Making a correlation heatmap.<br>

```python
sns.heatmap(data.corr(), annot=True, annot_kws={"size": 8}, fmt=".2f", linewidths=0.5)
```

The most impactful features are `Sex`, `Pclass` and `Fare`.<br>
Women were much more likely to survive, and so did the people with higher class and fare.
